```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# generiere eine Beispielzeitreihe mit 1000 Punkten im 3-dimensionalen Raum
t = np.linspace(0, 100, 1000)
x = np.sin(t) + 0.5*np.random.rand(1000)
y = np.cos(t) + 0.5*np.random.rand(1000)
z = t + 0.5*np.random.rand(1000)

# berechne die Distanzen zwischen allen Paaren von Punkten
dx = np.subtract.outer(x, x)
dy = np.subtract.outer(y, y)
dz = np.subtract.outer(z, z)
dist = np.sqrt(dx**2 + dy**2 + dz**2)

# wähle eine Reihe von Boxgrößen
box_sizes = np.logspace(np.log10(0.001), np.log10(0.1), 50)

# berechne die Anzahl der Boxen mit mindestens einem Punkt
num_boxes = []
for size in box_sizes:
    num_boxes.append(np.sum(dist < size))





# plotten Sie die Anzahl der Boxen als Funktion der Boxgröße
plt.loglog(box_sizes, num_boxes, 'o')

# schätzen Sie die Korrelationsdimension durch die lineare Regression
slope, intercept = np.polyfit(np.log(box_sizes), np.log(num_boxes), 1)
print("Korrelationsdimension: ", -slope)

# plotten Sie die lineare Regression
plt.loglog(box_sizes, np.exp(intercept)*box_sizes**slope, 'r-')

plt.xlabel('Boxgröße')
plt.ylabel('Anzahl der Boxen')
plt.show()

```
